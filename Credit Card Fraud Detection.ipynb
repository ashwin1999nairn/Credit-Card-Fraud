```python
from imblearn.over_sampling import SMOTE
```

```python
x=df.drop('Class',axis='columns')
y=df['Class']
```

```python
smote=SMOTE(sampling_strategy='minority')
x_sm,y_sm=smote.fit_resample(x,y)
y_sm.value_counts()
```

```python
## Feature Selection
model=ExtraTreesClassifier()
model.fit(x_sm,y_sm)
```

```python
model.feature_importances_
```

```python
plt.figure(figsize=(10,10))
feat=pd.Series(model.feature_importances_,index=x_sm.columns)
feat.nlargest(18).plot(kind='bar')
```

```python
plot=feat.nlargest(18)
```

```python
plot.index
```

```python
cols=['V12', 'V14', 'V11', 'V4', 'V17', 'V3', 'V16', 'V10', 'V9', 'V18', 'V1',
       'V2', 'V7', 'V19', 'V6', 'Time', 'V21', 'V24']
```

```python
x_new=x_sm[cols]
```

```python
x_sm.shape
```

```python
x_new.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y_sm,test_size=0.2,random_state=42)
```

```python
randomf=RandomForestClassifier()
```

```python
randomf.fit(x_train,y_train)
```

```python
y_pred=randomf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiondf=DecisionTreeClassifier()
decisiondf.fit(x_train,y_train)
y_pred=decisiondf.predict(x_test)
accuracy_score(y_test,y_pred)
```
